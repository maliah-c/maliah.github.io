```python
#this one works for me

import sqlite3
from sqlite3 import connect
import pandas as pd

df_projects = pd.read_excel("/projects.xlsx")
df_organizations = pd.read_excel("/participants.xlsx")
df_countries = pd.read_excel("/countries.xlsx")

con= sqlite3.connect("ecsel_database.db")
df_projects.to_sql("Projects", con, if_exists="replace", index = False) #by using that replace the last one if it is the second time runing it
df_organizations.to_sql("organizations", con, if_exists="replace", index = False)
df_countries.to_sql("countries", con, if_exists="replace", index = False)
con.close()

df_projects.head()
```
